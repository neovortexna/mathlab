```python
# Simulasi Pemodelan Pola Bunga Menggunakan Analisis Galat


import numpy as np
import matplotlib.pyplot as plt

print("=" * 80)
print("             SIMULASI KOMPUTASI NUMERIK: PEMODELAN POLA BUNGA             ")
print("=========================================================================\n")

sudut_emas_eksak = 137.507764
jumlah_biji = 300


sudut_hampiran = 137.500000


galat_absolut = abs(sudut_emas_eksak - sudut_hampiran)
galat_relatif = (galat_absolut / sudut_emas_eksak) * 100

print(f"--- ANALISIS GALAT SUDUT PERTUMBUHAN ---")
print(f"Nilai Eksak Sudut Emas : {sudut_emas_eksak:.6f} derajat")
print(f"Nilai Hampiran Numerik : {sudut_hampiran:.6f} derajat")
print(f"Galat Absolut (Ee)      : {galat_absolut:.6f} derajat")
print(f"Galat Relatif (Ee %)    : {galat_relatif:.6f}%\n")


n = np.arange(1, jumlah_biji + 1)
theta_galat = n * np.radians(sudut_hampiran)


r = np.sqrt(n)
x_galat = r * np.cos(theta_galat)
y_galat = r * np.sin(theta_galat)


plt.figure(figsize=(8, 8), facecolor='white')

plt.scatter(x_galat, y_galat, c=n, cmap='winter', edgecolors='black', s=55)


plt.title(f'Visualisasi Pola Bunga Matahari Akibat Galat Pembulatan\n(Galat Relatif: {galat_relatif:.4f}%)',
          fontsize=12, fontweight='bold', color='darkred', pad=15)
plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.5)


plt.axhline(0, color='black', linewidth=0.8, alpha=0.7)
plt.axvline(0, color='black', linewidth=0.8, alpha=0.7)

plt.xlabel('Koordinat X (Posisi Horisontal Kelopak)', fontsize=10)
plt.ylabel('Koordinat Y (Posisi Vertikal Kelopak)', fontsize=10)

plt.tight_layout()
plt.show()
```
